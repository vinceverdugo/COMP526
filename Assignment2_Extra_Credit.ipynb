```python
import numpy as np
import matplotlib.pyplot as plt

#Define f(x) function
def f(x):
    return np.exp(-x)*(3.2*np.sin(x)-0.5*np.cos(x))-3

#Define f'(x) using numerical differentiation
def fprime(x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

#Define kappa function
def kappa(g, gprime, x):
    return abs(gprime(x))*(abs(x)/abs(g(x)))

#Define bisection function
def bisection_hist(f, a, b, tol):
    hist = []
    while (abs(b-a) > tol):
        mid = (b+a) / 2
        hist.append(mid)
        if ((f(a)*f(mid)) < 0): #Code was working incorrectly with f(a)*f(b), but works correctly with f(a)*f(mid)
            b = mid
        else:
            a = mid
    return hist
```

```python
#Define intervals from (1.a) and (1.b)
x1 = np.linspace(-4,-2,1000)
x2 = np.linspace(2,4,1000)

#Title of graph
title = 'Plot of $\kappa \ for \ f(x) = e^{x}(3.2sin(x)-0.5cos(x)) - 3$ \n Over Two Intervals'

#Graph kappa over the two intervals
plt.subplot(2,1,1)
plt.plot(x1, kappa(f,fprime, x1), 'r', label=r'$\kappa \in [-4,-2]$')
plt.legend()
plt.title(title, size=10)

plt.subplot(2,1,2)
plt.plot(x2, kappa(f,fprime, x2), 'b', label='$\kappa \in [2,4]$')
plt.legend()


#Plot and save as png
plt.savefig("COMP526_Assigment2EC_Graph.png")
plt.show()
```

By looking at the two graphs, our function $f(x) = e^{x}(3.2sin(x)-0.5cos(x)) - 3$ is ill-conditioned over the interval $[-4,-2]$ and well conditioned over $[2,4]$.
